## Part 1: Canids

```python
# open the file (that was cleaned in the shell)
infile = "clean-pbdb.csv"
with open(infile, 'r') as ff:
    canid_recs = ff.readlines()[1:]
```

### sorting the data

```python
def sort_name(data):
    return(data.split(",")[5])
    
    
another_csv = sorted(canid_recs, key = sort_name)   

output_file = open("sorted_canids.csv", "w")
for entry in another_csv:
    output_file.write(entry)
    
output_file.close()
```

### First function that populates a dictionary with ages

I have also added in the step to remove the quotes into the function

```python
def canid_dict(csv):
#use defaultdict
    from collections import defaultdict
    
    with open(csv, "r") as aa:
        csv_data = aa.readlines()
    
    
    species_ranges = defaultdict(list)


# make a dictionary
    for line in csv_data:
        line = line.replace('"', '')
        species = line.split(",")[5]
        if len(species.split(" ")) > 1:
            minage = float(line.split(',')[10])
            maxage = float(line.split(',')[11])
            meanage = (minage + maxage) / 2
    
            species_ranges[species].append(meanage) # add species as keys and mean ages as elements to the list of ages
    return species_ranges
    #print(meanage)
```

```python
new_dict = canid_dict("sorted_canids.csv")
new_dict
```

### Second function that spits out csv with genus, species, min age, max age

```python
def canid_summary(dictionary, outputfile):
    output = open(str(outputfile), "w")
    for key in sorted(dictionary.keys()): 
        ages = dictionary[key]
        maxage = max(ages)
        minage = min(ages)
        genus = key.split(" ")[0] 
        species =key.replace(" ", "_")  
        outline = "{},{},{},{}\n".format(genus, species, minage, maxage)
        output.write(outline)
    output.close()



```

```python
canid_summary(new_dict, "canid-ranges.csv")
    
```

### Third function that gives longest fossil range

```python
with open("canid-ranges.csv", "r") as jj:
    canid_ranges = jj.readlines()
rangedict = {}
for line in canid_ranges:
    element = line.split(",")
    species = element[1]
    fossilrange = float(element[3])-float(element[2])
    rangedict[species] = fossilrange
```

```python
rangedict
```

```python
def longest_range(dictionary):
    longest_species = max(dictionary, key=dictionary.get)
    return longest_species
```

```python
longest_range(rangedict)
```

### sorting the files

### Writing first function to populate dictionary with ages

```python
def rats_dict(csv):
#use defaultdict
    from collections import defaultdict
    
    with open(csv, "r") as bb:
        rats_data = bb.readlines()
    
    
    species_ranges = defaultdict(list)


# make a dictionary
    for line in rats_data:
        line = line.replace('"', '')
        species = line.split(",")[5]
        if len(species.split(" ")) > 1:
            minage = float(line.split(',')[10])
            maxage = float(line.split(',')[11])
            meanage = (minage + maxage) / 2
    
            species_ranges[species].append(meanage) # add species as keys and mean ages as elements to the list of ages
    return species_ranges
    #print(meanage)
```

```python
rat_new_dict = rats_dict("sorted_rats.csv")
rat_new_dict
```

###  Second function that spits out csv with genus, species, min age, max age

```python
def rat_summary(dictionary, outputfile):
    output = open(str(outputfile), "w")
    for key in sorted(dictionary.keys()): 
        ages = dictionary[key]
        maxage = max(ages)
        minage = min(ages)
        genus = key.split(" ")[0] 
        species =key.replace(" ", "_")  
        outline = "{},{},{},{}\n".format(genus, species, minage, maxage)
        output.write(outline)
    output.close()
```

```python
rat_summary(rat_new_dict, "rat-ranges.csv")
```
